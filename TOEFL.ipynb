```python
import nltk
nltk.download('gutenberg')
nltk.download('punkt')
toto = open("TOTO.txt", "r")
toto = nltk.corpus.gutenberg.words(toto)
```

**Total of words**

```python
print(len(toto))
```

**Total of distinct words**

```python
print(len(set(toto)))
```

```python
toto_dist=(set(toto))
```

**Total of repeated words**


```python
print(len(toto)-len(toto_dist))
```

**Words and frequencies**


```python
toto_freq = nltk.FreqDist(toto)
toto_freq
toto_freq.most_common()
```
